```python
import pandas as pd

df = pd.read_csv('fine_food_reviews_1k.csv')
df
```

```python
#  토큰 수 확인 

import tiktoken

gpt4o_encoding = tiktoken.encoding_for_model("gpt-4o")

df['n_tokens']=df['Text'].apply(lambda x: len(gpt4o_encoding.encode(x)))
df[['Text', 'n_tokens']].head()
```

```python
df['n_tokens'].describe()
```

```python
def text_to_embedding(texts):
    texts = [text.replace("\n", " ") for text in texts]

    response=client.embeddings.create(
      model="text-embedding-3-small",
      input=texts
    )
    return [data.embedding for data in response.data]
```

```python
# 임베딩 처리 
df['embeddings']=text_to_embedding(df['Text'].tolist())

```

### 코사인 유사도 측적

```python
import numpy as np

cos_sim = lambda a, b: np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))   

def get_similar_texts(query, embed_df, top_n=5):
    query_embed = text_to_embedding([query])[0]

    embed_df['cos_sim'] = embed_df['embeddings'].apply(lambda x: cos_sim(x, query_embed))
    return embed_df.sort_values('cos_sim', ascending=False).head(top_n) # ascending=False -> 유사도가 높은 것들 위로


```

```python
import numpy as np

# 코사인 유사도 함수
cos_sim = lambda a, b: np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def get_similar_texts(query, embed_df, top_n=5):
    # query를 문자열로 받음
    query_embed = text_to_embedding([query])[0]

    # 유사도 계산
    embed_df['cos_sim'] = embed_df['embeddings'].apply(lambda x: cos_sim(x, query_embed))

    # 상위 top_n 반환
    return embed_df.sort_values('cos_sim', ascending=False).head(top_n)

# 실행
get_similar_texts("coffee", embed_df)
```
